# Cara mengambil data dari API Covid19
Data yang saya gunakan berasal dari website https://apicovid19indonesia-v2.vercel.app/

Dokumentasi lengkapnya bisa baca disini: https://github.com/Reynadi531/api-covid19-indonesia-v2

### Import Semua Library yang dipakai

```python
import requests
import json
import pandas as pd
import datetime
from datetime import datetime
```

### Proses pengambilan data melalui API

```python
url = "https://apicovid19indonesia-v2.vercel.app/api/indonesia/harian"

payload={}
headers = {}

response = requests.request("GET", url, headers=headers, data=payload)

print(response.status_code)
print('Data berhasil diambil')
#print(response.text)
```

### Mengubah json file menjadi string

```python
text = json.dumps(response.json(), sort_keys=False, indent=4)
#print(text)
```

### Membuat dataframe

```python
df = pd.read_json(text, orient='records')
df
```

### Mengubah kolom tanggal&waktu menjadi tanggal

```python
#datetime to date conversion
df['tanggal'] = pd.to_datetime(df['tanggal']).dt.date
df['tanggal'] = pd.to_datetime(df['tanggal'])
df
```

### Slicing data 3 bulan terakhir(2021-12-20 sampai 2022-03-20)

```python
start_date = '2021-12-19'
end_date = '2022-03-20'

mask = (df['tanggal'] > start_date) & (df['tanggal'] <= end_date)

df = df.loc[mask]
df
```

### Simpan sebagai file csv

```python
df.to_csv('Indonesia_Covid_Update_20des21_20mar22.csv')
```
