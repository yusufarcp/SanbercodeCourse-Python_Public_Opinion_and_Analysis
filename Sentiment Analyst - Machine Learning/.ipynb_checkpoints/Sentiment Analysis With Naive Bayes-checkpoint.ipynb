# Data Preprocessing (Cleaning Word)

```python
# Define Preprocessing (case folding, stopwords, Slangword)
def case_folding(data:list=[]):
    result = []
    for teks in data:
        teks = teks.lower()
        temp = re.compile("["
                u"\U0001F600-\U0001F64F"  # emoticons
                u"\U0001F300-\U0001F5FF"  # symbols & pictographs
                u"\U0001F680-\U0001F6FF"  # transport & map symbols
                u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                           "]+", flags=re.UNICODE)
        teks = temp.sub(r'', teks)
        teks = re.sub(r'\d+', "", teks)
        teks = re.sub(r"http\S+", "", teks)
        teks = teks.translate(str.maketrans("","",string.punctuation))
        teks = teks.strip()
        result.append(teks)
        
    return result

def remove_stopwords(data:list=[]):
    result = []
    stopword = stopwords.words('indonesian')
    for sentence in data:
        sentences = [word for word in sentence if word not in stopword]
        result.append(' '.join(sentences))
    
    return result    

def tokenizing(data:list=[]):
    result = [word_tokenize(sentence) for sentence in data]
    return result

def slang_cleaner(data, slang_list:list=[]):
    #membuat fungsi yang gunanya untuk menghilangkan kata slang/gaul dari data
    #fungsi ini mempunyai dua input parameter, data:list dan slang:list(pasangan tuple)
    #output adalah listening
    result = []
    for tweet in data:
        # print(tweet)
        word_list = []       
        for word in word_tokenize(tweet):
            if slang_list.get(word.lower()):
                word = slang_list[word.lower()]
            word_list.append(word)
        result.append(' '.join(word_list))
    
    return result
```

```python
#Lakukan Text Preprocessing pada Data Training

slang_list = {
        'yg' : "yang",
        'gue': 'saya',
        'engga': 'tidak',
        'lu' : 'kamu',
        'gak' : 'tidak',
        '<USERNAME>' : ' ',
        'username' : ' ',
        'nya' : ' ',
        'ga' : 'tidak',
        'aja' : 'saja',
        'org' : 'orang',
        'bgt' : 'banget',
        'jd' : 'jadi',
        'udah' : 'sudah',
        'kalo' : 'kalau',
        'klo' : 'kalau',
        'sm' : 'sama',
        'si' : ' ',
        'lo' : 'kamu',
        'sih' : ' ',
        'mbak' : 'mba',
        'jg' : 'juga',
        'udh' : 'sudah',
        'n' : ' ',
        'pake' : 'pakai',
        'deh' : ' '
    }
# slang: username,nya, ga, aja, org, bgt, jd, udah, kalo, si,klo,sm, lo, sih,mbak
cuitan = df_tweets_model['Text']
lowercase = case_folding(cuitan)
slangword = slang_cleaner(lowercase, slang_list)
clean_texts = remove_stopwords(tokenizing(slangword))
clean_text = pd.DataFrame(clean_texts, columns =['text_cleaning'])
df_tweets_model = pd.concat([df_tweets_model,clean_text], axis=1, join='outer')
df_tweets_model
```

# Machine Learning dan Data Teks
Disini merupakan proses utama setelah dilakukan pembersihan data. Algoritma **machine learning** yang digunakan yaitu ***Naive Bayes***.

```python
vectorizer = TfidfVectorizer (max_features=2500)
model_g = GaussianNB()
```

```python
v_data = vectorizer.fit_transform(df_tweets_model['text_cleaning']).toarray()
vectorizer.fit_transform(df_tweets_model['text_cleaning'])
```

### Train-Test Split

Pada **Data Training** akan dilakukan proses ***train-test split*** untuk mengetahui keakuratan dari kualitas algoritma yang akan dipakai. Proses ***train-test split*** sendiri akan membagi **Data Training** dengan komposisi ***80% Training*** dan ***20% Testing****.

```python
X_train, X_test, y_train, y_test = train_test_split(v_data, df_tweets_model['Sentiment'], test_size=0.2, random_state=18)

model_g.fit(X_train,y_train)
```

### Prediction and Report
Setelah proses pelatihan algoritma *machine learning* terhadap ***Training***, proses selanjutnya yaitu mencoba untuk melakukan prediksi terhadap ***Testing*** dalam **Data Training** serta mengetahui nilai akurasi dari proses prediksi tersebut.

```python
y_preds = model_g.predict(X_test)

print(confusion_matrix(y_test,y_preds))
print(classification_report(y_test,y_preds))
print('nilai akurasinya adalah ',accuracy_score(y_test, y_preds))
```

Nilai akurasi menunjukan nilai yang cukup baik untuk sebuah performa dari algorima machine learning. Untuk acuan nilai akurasi yang baik dapat melihat refensi lain di internet. Proses selanjutnya kita akan terapkan algoritma machine learning ini pada **Data Test**.

### Do predict for testing data!

```python
v_data2 = vectorizer.transform(df_tweets["tweet_clean"]).toarray()
y_classification = model_g.predict(v_data2)
```

```python
df_tweets["sentiment"] = list(y_classification)
```

# Get Know What Word and topics in positive Sentiment

Pada tahap ini kita dapat mengetahui *tweet* apa saja yang termasuk ke dalam ***Sentimen Positif***. Kita juga dapat mengetahui *kata* apa saja yang sering muncul dalam ****Sentimen Positif*** berdasarkan analisis distribusi **bi-gram** dan **wordcloud**. Hal ini dapat membantu kita untuk menganalisis **Topik Utama** dalam data yang kita miliki.

```python
posval = pd.DataFrame(df_tweets.loc[df_tweets['sentiment'] == 1, 'tweet_clean'])
posval
```

```python
#Pemisahan Kata/tokenize
posval['tweet_clean'] = posval['tweet_clean'].apply(lambda x: word_tokenize(str(x)))
ptweets = [word for tweet in posval['tweet_clean'] for word in tweet]

# Distribusi bi-gram
presult = pd.Series(nltk.ngrams(ptweets, 2)).value_counts()[:20]
print (presult)
```

Dari hasil analisis ***bi-gram*** dapat diketahui bahwa kata **'jawa barat'** dan **'umk jawa'** adalah kata2 yang paling sering muncul. Artinya adalah data tweet yang kita miliki, lebih banyak membahas tentang umk di pulau jawa, khususnya jawa bagian barat jika dilihat dari sudut pandang sentimen positif.

```python
# WordCloud
wordcloud = WordCloud(width=1600, height=800, max_font_size=200, background_color='black')
wordcloud.generate(str(ptweets))
```

```python
plt.figure(figsize=(12,10))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

Dari hasil ***wordcloud*** dapat diketahui juga bahwa **'umk, jawa, barat dan buruh'** menjadi kata yang paling sering muncul. Oleh karena itu berdasarkan sentimen analisis bernilai positif menunjukan topik yang dibahas adalah terkait ***umk buruh di jawa barat*** .
